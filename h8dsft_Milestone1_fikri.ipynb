Alasan menggunakan analisis pearson yaitu:
1. Mengukur hubungan linier antara dua variabel numerik.

##### Melakukan korelasi dengan Pearson dan divisualisasikan

````python
import seaborn as sns

# Default Korelasi Pearson
corr = df.corr()
f, ax = plt.subplots(figsize=(10, 8))
color = plt.get_cmap('YlGnBu')
color.set_bad('lightblue')
sns.heatmap(corr, annot=True, fmt=".2f", cmap=color)
plt.show()
````

````python
rank = corr[['GDP_current_US']]
sns.heatmap(rank, annot=True, fmt=".2f", cmap=color)
````

Dari gambar di atas didapatkan bahwa untuk menganalisis kolom `GDP_current_US` terdapat 5 kolom yang berkorelasi dengan nilai diatas 0.3 yaitu kolom `agricultural_land`, `land_area`, `life_expectancy_at_birth`, `population`, dan `rural_population`

##### Mencoba menggunakan korelasi yang lain seperti (Spearman dan Kendall)

- Korelasi antara GDP dan Agriculture

````python
from scipy import stats

corr_r, pval_p = stats.pearsonr(df['GDP_current_US'], df['agricultural_land'])
corr_rho, pval_s = stats.spearmanr(df['GDP_current_US'], df['agricultural_land'])
corr_tau, pval_k = stats.kendalltau(df['GDP_current_US'], df['agricultural_land'])

print(f"r-correlation: {corr_r:.2f}, p-value: {pval_p}")
print(f"rho-correlation: {corr_rho:.2f}, p-value: {pval_s}")
print(f"tau-correlation: {corr_tau:.2f}, p-value: {pval_k}")
````

Dari hasil analisis korelasi antara `GDP_current_US` (Gross Domestic Product dalam dolar Amerika Serikat) dan `agricultural_land` (luas lahan pertanian) menggunakan berbagai metrik korelasi, ditemukan hal-hal berikut:

1. Korelasi Pearson (r-correlation): Hasil korelasi Pearson adalah sekitar 0.35, yang menunjukkan adanya hubungan positif antara `GDP_current_US` dan `agricultural_land`. Artinya, ketika `GDP` suatu negara naik, cenderung juga terjadi peningkatan luas lahan pertanian. P-value yang sangat rendah (kurang dari 0.001) menunjukkan bahwa hubungan ini sangat signifikan secara statistik.

2. Korelasi Spearman (rho-correlation): Korelasi Spearman adalah sekitar 0.46, yang menunjukkan korelasi yang lebih kuat antara kedua variabel ini. Hasil ini mengindikasikan bahwa hubungan antara `GDP_current_US` dan `agricultural_land` adalah positif dan kuat. P-value yang sangat rendah menegaskan signifikansinya.

3. Korelasi Kendall (tau-correlation): Korelasi Kendall adalah sekitar 0.32, menunjukkan adanya hubungan positif antara kedua variabel tersebut. Meskipun kurang kuat dibandingkan dengan Spearman, tetapi tetap signifikan. P-value yang rendah menunjukkan bahwa hubungan ini juga memiliki signifikansi statistik.

Kesimpulan dari hasil analisis ini adalah bahwa ada hubungan positif antara `GDP_current_US` dan `agricultural_land`. Semakin besar GDP suatu negara, semakin besar luas lahan pertanian yang cenderung ada. Korelasi Spearman menunjukkan hubungan yang lebih kuat dan lebih konsisten dibandingkan dengan metode lainnya, seperti Pearson dan Kendall. Kedua variabel ini memiliki signifikansi statistik yang tinggi, menunjukkan bahwa hubungan ini bukanlah kebetulan semata.

### 3. Kapan GDP pada setiap negara di Asia Tenggara mencapai titik tertinggi? (Visualisasi Biasa)

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Filter data hanya untuk 20 tahun terakhir
last_20_years = df[df['date'] >= '2003-01-01']

# Inisialisasi sebuah dictionary untuk menyimpan tahun puncak GDP dan nilai GDP untuk setiap negara
peak_data = {}

# Loop melalui setiap negara di data
for country in last_20_years['country'].unique():
    data_country = last_20_years[last_20_years['country'] == country]
    peak_gdp = data_country['GDP_current_US'].max()
    peak_year = data_country[data_country['GDP_current_US'] == peak_gdp]['date'].values[0]
    peak_data[country] = (peak_year, peak_gdp)

# Sorting dictionary berdasarkan nilai GDP
sorted_peak_data = dict(sorted(peak_data.items(), key=lambda item: item[1][1], reverse=True))

# Menyiapkan data untuk visualisasi
countries = list(sorted_peak_data.keys())
years_and_gdp = list(sorted_peak_data.values())
peak_years = [item[0] for item in years_and_gdp]
peak_gdp = [item[1] for item in years_and_gdp]
colors = np.random.rand(len(countries), 3)

# Membuat diagram batang horizontal
plt.figure(figsize=(10, 8))
plt.barh(countries, peak_gdp, color=colors)
plt.xlabel('Nilai GDP')
plt.ylabel('Negara')

# Menambahkan tahun puncak GDP sebagai teks di dalam diagram
for i, country in enumerate(countries):
    # print(i)
    plt.text(peak_gdp[i], i, f'Tahun {peak_years[i][:4]}', va='center', fontsize=9)

plt.title('Nilai GDP Puncak di Negara-Negara Asia Tenggara (20 Tahun Terakhir)')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()
````

### 7. Antara singapura dan malaysia mana yang lebih baik pertumbuhan GDPnya? (Statistik Inferensial)

Two-Sample Independent Test

H0 : GDP pertahun Malaysia == Singapura

H1 : GDP pertahun Malaysia != Singapura

````python
#Year GDP
ML = df[df['country']=='Malaysia'][['year','GDP_current_US']].groupby(['year']).sum()
SG = df[df['country']=='Singapore'][['year','GDP_current_US']].groupby(['year']).sum()
````

````python
print('Average Year GDP_current_US of Malaysia =',ML['GDP_current_US'].mean())
print('Average Year GDP_current_US of Singapore =',SG['GDP_current_US'].mean())
````

````python
t_stat, p_val = stats.ttest_ind(ML['GDP_current_US'],SG['GDP_current_US'])
print('T-Statistic:',t_stat)
print('P-value:',p_val)
````

#### Jawaban No.7

Untuk menafsirkan hasil dari uji statistik ini, dapat menggunakan nilai T-Statistic dan P-value. Hipotesis yang diajukan adalah:

**H0 (Hipotesis Nol)**: Rata-rata GDP per tahun di Malaysia sama dengan rata-rata GDP per tahun di Singapura.

**H1 (Hipotesis Alternatif)**: Rata-rata GDP per tahun di Malaysia tidak sama dengan rata-rata GDP per tahun di Singapura.

Kemudian, hasil uji statistik memberikan:

- T-Statistic: 0.13797686375181756
- P-value: 0.8904825092049504

Dalam konteks uji t ini, kita biasanya menggunakan tingkat signifikansi (alpha) untuk mengevaluasi P-value. Sebagai contoh, jika alpha adalah 0.05, maka kita akan memutuskan apakah menerima atau menolak hipotesis nol berdasarkan P-value:

- Jika P-value lebih kecil dari alpha (0.05), kita akan menolak hipotesis nol (H0) dan menganggap ada perbedaan signifikan antara rata-rata GDP Malaysia dan Singapura.
- Jika P-value lebih besar dari alpha (0.05), kita tidak memiliki cukup bukti statistik untuk menolak hipotesis nol, sehingga kita tidak dapat menyimpulkan adanya perbedaan signifikan antara rata-rata GDP kedua negara.

Dalam kasus ini, P-value adalah 0.8904825092049504, yang jauh lebih besar dari alpha (0.05). Oleh karena itu, kita tidak memiliki cukup bukti statistik untuk menolak hipotesis nol. Dengan kata lain, berdasarkan uji statistik ini, tidak ada bukti yang cukup untuk menyatakan bahwa rata-rata GDP per tahun di Malaysia dan Singapura berbeda secara signifikan.
